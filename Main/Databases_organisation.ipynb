```python
import pandas as pd
from datetime import date, datetime, timedelta, time
```

### Events

```python
events = pd.read_excel("Events_database_final.xlsx", index_col=0, na_values=["Не вказано", "Розпродано"])
```

```python
months_dict = {"січ.": "01","лют.": "02", "бер.": "03", "квіт.": "04", "трав.": "05", "черв.":"06", "лип.":"07", "серп.":"08", "вер.":"09", "жовт.":"10", "лист.":"11", "груд.":"12"}
events["Start time"] = events.apply(lambda row: row["Date"][:2] + "." + months_dict[row["Date"][3:].strip()] + ".2023" + " " + row["Start time"], axis="columns").map(lambda x: datetime.strptime(x, "%d.%m.%Y %H:%M"))
events.drop("Date", inplace = True, axis="columns")
events.sort_values(by="Start time")
mean_dur = round(events["Duration"].dropna().mean())
events["Duration"] = events["Duration"].fillna(mean_dur).astype("int64")
```

```python
events["Prices lower"] = events["Price range ₴"].map(lambda x: x[:x.find("-")].strip() if "-" in x else x)
events["Prices upper"] = events["Price range ₴"].map(lambda x: x[x.find("-") + 1:].strip() if "-" in x else x)
events.drop("Price range ₴", inplace=True,axis="columns")
events = events.astype({"Prices lower": "int32", "Prices upper": "int32"})
```

```python
events["End time"] = events.apply(lambda row: row["Start time"] + timedelta(minutes=row["Duration"]), axis="columns")
events.sort_values(by="Start time").head(5)
```

```python
even = events.copy()
even.insert(11, "Hashtag", even["Hashtags"])
del even['Hashtags']
even.sort_values(by="Start time").head(5)
```

```python
events.to_csv("Events.csv")
```

### Restaurants

```python
restaurants = pd.read_excel("database_restaurant_final.xlsx", index_col=0, na_values = "Closed").replace("Open24hours", "00:00AM–12:00PM")
menu = pd.read_excel("database_restaurant_final.xlsx", sheet_name=1, index_col=0)
```

```python
closed = list(restaurants.loc[restaurants["work_hours"].isnull()].index)
menu = menu.loc[~menu["ID_Rest"].isin(closed)]
restaurants = restaurants.dropna().rename(columns={"location_address": "Address", "location_coordinates":"Coordinates"})
```

```python
def format_time(time):
    if "12" in time:
        return datetime.strptime("23:59", "%H:%M").time() if "AM" in time else datetime.strptime("12:00", "%H:%M").time()
    if "AM" in time:
        return datetime.strptime(time[:time.find("AM")].strip(), "%H:%M").time()
    elif "PM" in time:
        return (datetime.strptime(time[:time.find("PM")].strip(), "%H:%M") + timedelta(hours=12)).time()
    else:
        return datetime.strptime(time, "%H:%M").time()
    
restaurants["Open time"] = restaurants["work_hours"].map(lambda x: format_time(x[:x.find("–")]))
restaurants["Close time"] = restaurants["work_hours"].map(lambda x: format_time(x[x.find("–") + 1:]))
restaurants.drop("work_hours", inplace=True, axis='columns')
```

```python
menu["Price"] = menu["Price"].map(lambda x: float(x.replace("\xa0","").replace(",",".").replace("₴","")))
```

```python
restaurants["Average price"] = menu.groupby(["ID_Rest"]).apply(lambda df: round(df[(df["Price"] >= df["Price"].describe().iloc[4]) & (df["Price"] <= df["Price"].describe().iloc[6])]["Price"].mean()))
quart = restaurants["Average price"].describe()
restaurants["Overall price level"] = restaurants["Average price"].map(lambda level: "Low" if level < quart.iloc[4] else "Medium-Low" if level < quart.iloc[5] else "Medium-High" if level < quart.iloc[6] else "High")
```
